```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# Coordonnées de Rennes (Bretagne)
rennes_lat, rennes_lon = 48.1173, -1.6778

# Créer la carte centrée sur Rennes (Bretagne)
carte_bretagne = folium.Map(location=[rennes_lat, rennes_lon], zoom_start=8)

# Initialize the MarkerCluster object
marker_cluster = MarkerCluster().add_to(carte_bretagne)

# Sample 1000 rows randomly from the DataFrame
sampled_df = df.sample(n=1000, random_state=42)  # n=1000 to select 1000 random rows

# Add the markers with clustering for the sampled data
for idx, row in sampled_df.iterrows():
    if pd.notna(row['lat']) and pd.notna(row['lon']):
        # Create a popup text for each marker
        popup = f"Type de bâtiment: {row['Type_bâtiment']}<br>Surface habitable: {row['Surface_habitable_logement']} m²<br>Étiquette DPE : {row['Etiquette_DPE']}"
        
        # Create a marker for each property and add it to the marker cluster
        folium.Marker(
            location=[row['lat'], row['lon']],
            popup=popup
        ).add_to(marker_cluster)

# Display the map
# carte_bretagne
carte_bretagne
```

```python
import streamlit as st
import pandas as pd

# Titre de la page
st.title("Affichage des Données")

# Charger les données
data_path = "data/processed/full_data_bretagne.csv"
@st.cache_data
def load_data(path):
    return pd.read_csv(path, sep=";")

# Afficher les données avec une option de cacher
df = load_data(data_path)
if st.checkbox("Afficher toutes données"):
    # Afficher le dataframe de base
    st.dataframe(df)

# Récupérer les valeurs uniques de 'Type_bâtiment'
type_batiment_options = df['Type_bâtiment'].unique()

# Créer une liste vide pour stocker les types sélectionnés
selected_types = []

# Disposer les checkboxes côte à côte
st.write("Filtrer par type de bâtiment :")
columns = st.columns(len(type_batiment_options))  # Crée une colonne pour chaque type unique

for i, type_batiment in enumerate(type_batiment_options):
    # Ajouter chaque checkbox dans une colonne différente
    if columns[i].checkbox(type_batiment):
        selected_types.append(type_batiment)

# Filtrer le dataframe en fonction des types sélectionnés
if selected_types:
    df_filtered = df[df['Type_bâtiment'].isin(selected_types)]
else:
    df_filtered = df  # Si aucun type n'est sélectionné, on garde toutes les données

# Affichage des résultats filtrés
st.write("Données filtrées", df_filtered)

# Ajouter un bouton pour exporter les données filtrées en CSV
csv = df_filtered.to_csv(index=False).encode('utf-8')
st.download_button(
    label="Exporter les données filtrées en CSV",
    data=csv,
    file_name="données_filtrées.csv",
    mime="text/csv"
)
```
